```python
import pandas as pd

df = pd.read_parquet("../../data_lake/ml/datasets/raw_ml_dataset_new_version.parquet")

print("Linhas:", len(df))

print("\nMissing values")
print(df.isna().sum())

print("\nDuplicatas")
print(df.duplicated().sum())
```

```python
df_deduplicated = df.drop_duplicates()
print("\nLinhas após deduplicação:", len(df_deduplicated))
```

```python
print("\nMissing values")
print(df_deduplicated.isna().sum())
```

```python
df_deduplicated.dropna(inplace=True)
df_deduplicated.reset_index(drop=True)
print("\nLinhas após remoção de missing values:", len(df_deduplicated))
```

```python
import os

OUTPUT_DIR = "../../data_lake/ml/datasets"
os.makedirs(OUTPUT_DIR, exist_ok=True)

OUTPUT = f"{OUTPUT_DIR}/df_validation.parquet"

df_deduplicated.to_parquet(OUTPUT, index=False)
```
