```python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):

    R = 6371  # km

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)

    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1)
        * np.cos(lat2)
        * np.sin(dlon / 2) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))

    return R * c
```

```python
df["distance_km"] = haversine(
    df["customer_avg_lat"],
    df["customer_avg_lng"],
    df["seller_avg_lat"],
    df["seller_avg_lng"]
)
```
