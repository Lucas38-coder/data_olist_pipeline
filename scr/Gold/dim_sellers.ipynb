```python
import duckdb
import os
```

```python
SELLERS = "../../data_lake/silver/sellers/sellers_silver.parquet"

OUTPUT_DIR = "../../data_lake/gold/dimensions"
os.makedirs(OUTPUT_DIR, exist_ok=True)

OUTPUT = f"{OUTPUT_DIR}/dim_sellers.parquet"

con = duckdb.connect()
```

```python
con.execute(f"""
COPY (
    SELECT
        seller_id,
        zip_code_prefix,
        seller_city,
        seller_state
    FROM read_parquet('{SELLERS}')
)
TO '{OUTPUT}' (FORMAT PARQUET);
""")

print("dim_sellers criada.")
```
