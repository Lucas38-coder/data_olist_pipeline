```python
import duckdb
import os
```

```python
REVIEWS = "../../data_lake/silver/order_reviews/order_reviews_silver.parquet"

OUTPUT_DIR = "../../data_lake/gold/facts"
os.makedirs(OUTPUT_DIR, exist_ok=True)

OUTPUT = f"{OUTPUT_DIR}/fact_reviews.parquet"

con = duckdb.connect()
```

```python
con.execute(f"""
COPY (
    SELECT
        review_id,
        order_id,
        review_score,
        low_score_flag,
        review_creation_ts
    FROM read_parquet('{REVIEWS}')
)
TO '{OUTPUT}' (FORMAT PARQUET);
""")

print("fact_reviews criada.")
```
