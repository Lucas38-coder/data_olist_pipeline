```python
import duckdb
import os
```

```python
ORDERS = "../../data_lake/silver/orders/orders_silver.parquet"
ITEMS = "../../data_lake/silver/order_items/order_items_silver.parquet"
PAYMENTS = "../../data_lake/silver/order_payments/order_payments_silver.parquet"

OUTPUT_DIR = "../../data_lake/gold/facts"
os.makedirs(OUTPUT_DIR, exist_ok=True)

OUTPUT = f"{OUTPUT_DIR}/fact_orders.parquet"

con = duckdb.connect()
```

```python
con.execute(f"""
COPY (
    SELECT
        o.order_id,
        o.customer_id,
        i.product_id,
        i.seller_id,

        o.purchase_ts,
        o.purchase_date,

        i.price,
        i.freight_value,
        i.total_item_value,

        p.payment_value as total_payment_value,

        o.is_delivered,
        o.is_late_delivery,
        o.delivery_time_days

    FROM read_parquet('{ORDERS}') o

    LEFT JOIN read_parquet('{ITEMS}') i
    ON o.order_id = i.order_id

    LEFT JOIN read_parquet('{PAYMENTS}') p
    ON o.order_id = p.order_id
)
TO '{OUTPUT}' (FORMAT PARQUET);
""")

print("fact_orders criada.")
```
